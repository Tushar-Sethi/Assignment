```python
#Finding Population Gain in Percentages
array_F=[]
array_Income=[]
for i,j in df_Final.iterrows():
    array=[]
    array2=[]
    array.append(j['Country Name'])
    array.append((j['2020']-j['1960'])/j['1960'])
    array.append(j['2020'])
    array.append(j['Region'])
    array2.append(j['Country Name'])
    array2.append(j['IncomeGroup'])
    array_F.append(array)
    array_Income.append(array2)

array_F.pop()
array_Income.pop()
Df_Population_Gain = pd.DataFrame(array_F)
# Df_Population_Gain
Df_Population_Gain.rename({0:'Country Name',1:'Population Gain',2:'Current Population',3:'Region'},axis=1,inplace=True)
Df_Population_Gain.set_index('Country Name',inplace=True)
```

```python
array_Income
Df_Income = pd.DataFrame(array_Income)
Df_Income.rename({0:'Country Name',1:'Income Group'},axis=1,inplace=True)
Df_Income
```

```python
Regions = Df_sorted.Region.unique()
```

```python
Df_sorted
```

```python
dict_for_Region_Countries ={}
for i,j in Df_sorted.iterrows():
    for k in Regions:
        if(j['Region'] == k):
            if k in dict_for_Region_Countries.keys():
                dict_for_Region_Countries[k].append(i)
            else:
                dict_for_Region_Countries[k] = [i]
```

```python
pie_Array_Final=[]
for i in dict_for_Region_Countries:
    pie_Array=[]
#     print(i,len(dict_for_Region_Countries[i]))
    pie_Array.append(i)
    pie_Array.append(len(dict_for_Region_Countries[i]))
    pie_Array_Final.append(pie_Array)
# pie_Array_Final

pie_DF = pd.DataFrame(pie_Array_Final)
pie_DF.rename({0:'Regions',1:'No of Countries Included'},axis=1,inplace=True)
pie_DF
```
